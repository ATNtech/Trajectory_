```python
from bs4 import BeautifulSoup
with open("vkparser/Поиск сообществ.html") as fp:
    soup = BeautifulSoup(fp, 'html.parser')
#x = soup.find("div", {"id": "results"})
#x.find_all("div", {"class": "labeled title"})
x=[]
for data in soup.find_all("div", {"id": "results"}):
    for data2 in data.find_all("div", {"class": "labeled title"}):
        for a in data2.find_all('a'):
               x.append(a.get('href'))
y=[]                
for text1 in soup.find_all("div", {"id": "results"}):
    for text2 in text1.find_all("div", {"class": "labeled title"}):
#        print(text2.text)
        y.append(text2.text)

z=[]    
for img1 in soup.find_all("div", {"class": "search_item_img_stories"}):
    for img2 in img1.find_all('img'):
        z.append(img2.get('src'))    
#print(x)
#print(y)
#print(z)
```
